```python
import numpy as np
import csv
```

```python
deliveries = np.genfromtxt(
    "deliveries.csv",
    delimiter=",",
    dtype=str,
    skip_header=1
)

with open("matches.csv", "r", encoding="utf-8") as f:
    reader = csv.reader(f)
    header = next(reader)  
    data = [row for row in reader]

matches = np.array(data, dtype=str)
```

```python
match_ids = deliveries[:,0].astype(int)
batting_team = deliveries[:,2]
batter = deliveries[:,6]
bowler = deliveries[:,7]
batsman_runs = deliveries[:,9].astype(int)
over = deliveries[:,4].astype(int)
ball_raw = deliveries[:,5]
```

```python
mask_numeric = np.char.isdigit(ball_raw)
ball = ball_raw[mask_numeric].astype(int)
runs_for_valid_balls = batsman_runs[mask_numeric]
```

```python
unique_batters = np.unique(batter)
strike_rates = []

for player in unique_batters:
    player_runs = batsman_runs[batter==player].sum()
    player_balls = np.sum(mask_numeric & (batter==player))
    if player_balls > 0:
        sr = (player_runs / player_balls) * 100
        strike_rates.append([player, sr])

strike_rates = np.array(strike_rates, dtype=object)
```

# TASK-1

```python
unique_matches = np.unique(match_ids)
total_runs_per_match = np.array([
    [m_id, batsman_runs[match_ids==m_id].sum()]
    for m_id in unique_matches
])

print("Total Runs per Match:\n", total_runs_per_match[:10])
```

# TASK - 2

```python
unique_batters = np.unique(batter)
runs_per_batter = np.array([
    [player, batsman_runs[batter==player].sum()]
    for player in unique_batters
])

sorted_idx = np.argsort(runs_per_batter[:,1].astype(int))[::-1]
top5 = runs_per_batter[sorted_idx][:5]
print("\nTop 5 Batters:\n", top5)
```

# TASK-3

```python
balls_faced = np.array([
    [player, np.sum(batter==player)]
    for player in unique_batters
])

strike_rates = np.array([
    [player,
     ((batsman_runs[batter==player].sum() / balls_faced[balls_faced[:,0]==player][:,1].astype(int)[0]) * 100)]
    for player in unique_batters
])

print("\nSample Strike Rates:\n", strike_rates[40:80])
```

# TASK - 4

```python
unique_bowlers = np.unique(bowler)
economy_rates = np.array([
    [bw, batsman_runs[bowler==bw].sum() / (len(batsman_runs[bowler==bw]) / 6)]
    for bw in unique_bowlers
])

print("\nSample Economy Rates:\n", economy_rates[:10])
```

# TASK - 5

```python
avg_runs_per_over = np.array([
    batsman_runs[over==ov].mean()
    for ov in range(0,20)
])

print("\nAverage Runs per Over (0-19):")
for ov, val in enumerate(avg_runs_per_over):
    print(f"Over {ov+1}: {val:.2f}")
```

# TASK - 6

```python
fours = np.sum(batsman_runs==4)
sixes = np.sum(batsman_runs==6)

print("\nTotal Fours:", fours, "Total Sixes:", sixes)

teams = np.unique(batting_team)
team_boundaries = np.array([
    [team, np.sum((batting_team==team) & ((batsman_runs==4)|(batsman_runs==6)))]
    for team in teams
])
max_team = team_boundaries[np.argmax(team_boundaries[:,1].astype(int))]
print("Team with most boundaries:", max_team)
```

# TASK - 7

```python
death_mask = (over>=16) & (over<=20)
death_runs = batsman_runs[death_mask].sum()

death_team_runs = np.array([
    [team, batsman_runs[(batting_team==team)&death_mask].sum()]
    for team in teams
])
top_death_team = death_team_runs[np.argmax(death_team_runs[:,1].astype(int))]
print("\nDeath Overs Total Runs:", death_runs)
print("Top Death Overs Team:", top_death_team)
```

# TASK - 8

```python
highest_match = total_runs_per_match[np.argmax(total_runs_per_match[:,1].astype(int))]
print("\nHighest Scoring Match:", highest_match)
```

# TASK - 9

```python
match_team_runs = np.array([
    [mid, team, batsman_runs[(match_ids==mid)&(batting_team==team)].sum()]
    for mid in unique_matches for team in np.unique(batting_team[match_ids==mid])
])

winners = []
for mid in unique_matches:
    rows = match_team_runs[match_team_runs[:,0]==str(mid)]
    if len(rows)==2:
        winner = rows[np.argmax(rows[:,2].astype(int))]
        winners.append((mid, winner[1], winner[2]))
print("\nSample Winners:\n", winners[:10])
```

# TASK - 10

```python
match_ids_matches = matches[:,0].astype(int)

toss_winners = matches[:,7]  

toss_impact = []
for mid in unique_matches:
    if mid in match_ids_matches:
        toss_team = toss_winners[match_ids_matches==mid][0]
        toss_impact.append(str(toss_team)) 

print("\nSample Toss Impact :\n", toss_impact[:10])
```
