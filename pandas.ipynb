# STAGE 1

```python
import pandas as pd
```

```python
deliveries_df = pd.read_csv("deliveries.csv")
matches_df = pd.read_csv("matches.csv")
```

```python
print("Deliveries shape:", deliveries_df.shape)
print("Matches shape:", matches_df.shape)
```

```python
print("Deliveries columns:", deliveries_df.columns.tolist())
print("Matches columns:", matches_df.columns.tolist())
```

```python
print(deliveries_df.dtypes)
print(matches_df.dtypes)
```

# STAGE 2

```python
print("Missing values in deliveries_df:")
print(deliveries_df.isnull().sum())

print("\nMissing values in matches_df:")
print(matches_df.isnull().sum())
```

```python
matches_df['city'] = matches_df['city'].fillna('Unknown')
matches_df['method'] = matches_df['method'].fillna('Not Applicable')

deliveries_df['extras_type'] = deliveries_df['extras_type'].fillna('None')
deliveries_df['player_dismissed'] = deliveries_df['player_dismissed'].fillna('None')
deliveries_df['dismissal_kind'] = deliveries_df['dismissal_kind'].fillna('None')
deliveries_df['fielder'] = deliveries_df['fielder'].fillna('None')
```

```python
deliveries_match_ids = set(deliveries_df['match_id'])
matches_match_ids = set(matches_df['id'])

missing_in_matches = deliveries_match_ids - matches_match_ids
missing_in_deliveries = matches_match_ids - deliveries_match_ids

print("Match IDs in deliveries but not in matches:", missing_in_matches)
print("Match IDs in matches but not in deliveries:", missing_in_deliveries)
```

```python
print(deliveries_df.dtypes)
print(matches_df.dtypes)
```

```python
deliveries_df['match_id'] = deliveries_df['match_id'].astype(int)
deliveries_df['over'] = deliveries_df['over'].astype(int)
deliveries_df['ball'] = deliveries_df['ball'].astype(int)
deliveries_df['total_runs'] = deliveries_df['total_runs'].astype(int)

matches_df['id'] = matches_df['id'].astype(int)
```

```python
categorical_cols_deliveries = ['batting_team', 'bowling_team', 'batter', 'bowler']
for col in categorical_cols_deliveries:
    deliveries_df[col] = deliveries_df[col].astype('category')

categorical_cols_matches = ['team1', 'team2', 'winner', 'toss_winner']
for col in categorical_cols_matches:
    matches_df[col] = matches_df[col].astype('category')
```

```python
deliveries_df = deliveries_df.drop_duplicates()
matches_df = matches_df.drop_duplicates()
```

```python
print("Final Deliveries Shape:", deliveries_df.shape)
print("Final Matches Shape:", matches_df.shape)

print("\nAny missing values left (deliveries)?\n", deliveries_df.isnull().sum())
print("\nAny missing values left (matches)?\n", matches_df.isnull().sum())
```

# STAGE 3

```python
deliveries_df['computed_total_runs'] = (
    deliveries_df['batsman_runs'] + deliveries_df['extra_runs']
)

mismatch = deliveries_df[
    deliveries_df['computed_total_runs'] != deliveries_df['total_runs']
]

print("Mismatch rows:", len(mismatch))
```

```python
deliveries_df['is_boundary'] = deliveries_df['batsman_runs'].isin([4, 6]).astype(int)

deliveries_df['is_dot'] = (deliveries_df['total_runs'] == 0).astype(int)

deliveries_df['ball_id'] = deliveries_df['over'] * 100 + deliveries_df['ball']
```

```python
deliveries_df = deliveries_df.rename(columns={
    'match_id': 'match_id',
    'batsman': 'batter',  
})

matches_df = matches_df.rename(columns={
    'id': 'match_id'
})
```

```python
merged_df = deliveries_df.merge(
    matches_df,
    on='match_id',
    how='inner'
)
```

```python
print("Merged Shape:", merged_df.shape)
print("\nColumns:\n", merged_df.columns)
```

# STAGE 4

```python
total_runs_match = merged_df.groupby('match_id')['total_runs'].sum().reset_index()
```

```python
runs_per_team_match = (
    merged_df.groupby(['match_id', 'batting_team'])['total_runs']
    .sum()
    .reset_index()
)
```

```python
top_batters = (
    merged_df.groupby('batter')['batsman_runs']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)
```

```python
batter_stats = merged_df.groupby('batter').agg({
    'batsman_runs': 'sum',
    'ball': 'count'
}).rename(columns={'ball': 'balls_faced'})

batter_stats['strike_rate'] = (
    batter_stats['batsman_runs'] / batter_stats['balls_faced'] * 100
)
```

```python
bowler_stats = merged_df.groupby('bowler').agg({
    'total_runs': 'sum',
    'ball': 'count'
})

bowler_stats['overs'] = bowler_stats['ball'] / 6
bowler_stats['economy'] = bowler_stats['total_runs'] / bowler_stats['overs']

top_bowlers = bowler_stats.sort_values('economy').head(10)
```

```python
runs_per_match = merged_df.groupby(['batter', 'match_id'])['batsman_runs'].sum().reset_index()

consistency = runs_per_match.groupby('batter').agg({
    'batsman_runs': 'mean',
    'match_id': 'nunique'
}).rename(columns={
    'batsman_runs': 'avg_runs',
    'match_id': 'matches_played'
})

consistent_batters = consistency[
    consistency['matches_played'] > 30
].sort_values('avg_runs', ascending=False).head(10)
```

```python
highest_scores = (
    merged_df.groupby(['match_id', 'batter'])['batsman_runs']
    .sum()
    .reset_index()
    .sort_values('batsman_runs', ascending=False)
    .head(1)
)
```

```python
total_4s = (merged_df['batsman_runs'] == 4).sum()
total_6s = (merged_df['batsman_runs'] == 6).sum()

boundaries = merged_df[merged_df['batsman_runs'].isin([4, 6])]

top_boundary_players = (
    boundaries.groupby('batter')['batsman_runs']
    .count()
    .sort_values(ascending=False)
    .head(10)
)
```

```python
boundary_runs = merged_df[merged_df['batsman_runs'].isin([4, 6])]

boundary_stats = merged_df.groupby('batter')['batsman_runs'].sum()

boundary_contribution = (
    boundary_runs.groupby('batter')['batsman_runs'].sum() / boundary_stats * 100
).sort_values(ascending=False)
```

```python
dot_balls = merged_df[merged_df['total_runs'] == 0]

top_dot_bowlers = (
    dot_balls.groupby('bowler').size()
    .sort_values(ascending=False)
    .head(10)
)
```

```python
runs_per_over = (
    merged_df.groupby('over')['total_runs']
    .mean()
)
```

```python
powerplay = merged_df[merged_df['over'] <= 6]

powerplay_runs = (
    powerplay.groupby('batting_team')['total_runs']
    .sum()
    .sort_values(ascending=False)
)
```

```python
death_overs = merged_df[merged_df['over'] >= 16]

death_runs_team = (
    death_overs.groupby('batting_team')['total_runs']
    .sum()
    .sort_values(ascending=False)
)

death_runs_batter = (
    death_overs.groupby('batter')['batsman_runs']
    .sum()
    .sort_values(ascending=False)
)
```

```python
inning_runs = (
    merged_df.groupby(['match_id', 'inning'])['total_runs']
    .sum()
    .reset_index()
)
```

```python
toss_impact = merged_df.groupby('match_id').agg({
    'toss_winner': 'first',
    'winner': 'first'
})

toss_impact['toss_win_match_win'] = (
    toss_impact['toss_winner'] == toss_impact['winner']
)

toss_win_percentage = toss_impact['toss_win_match_win'].mean() * 100
```

```python
pom_runs = merged_df.groupby(['match_id', 'batter'])['batsman_runs'].sum().reset_index()

top_scorers = pom_runs.loc[pom_runs.groupby('match_id')['batsman_runs'].idxmax()]

comparison = top_scorers.merge(
    matches_df[['match_id', 'player_of_match']],
    left_on=['match_id', 'batter'],
    right_on=['match_id', 'player_of_match'],
    how='left'
)
```

```python
venue_stats = matches_df.groupby('venue').agg({
    'match_id': 'count'
}).rename(columns={'match_id': 'total_matches'})
```

```python
city_runs = merged_df.groupby('city')['total_runs'].mean()
```

```python
season_runs = merged_df.groupby('season')['total_runs'].sum()
```

```python
match_totals = merged_df.groupby(['match_id', 'batting_team'])['total_runs'].sum().reset_index()

predicted_winner = match_totals.loc[
    match_totals.groupby('match_id')['total_runs'].idxmax()
]

comparison = predicted_winner.merge(
    matches_df[['match_id', 'winner']],
    on='match_id'
)

accuracy = (comparison['batting_team'] == comparison['winner']).mean() * 100
```

# STAGE 6

```python
def standardize_columns(df):
    df.columns = (
        df.columns
        .str.strip()
        .str.lower()
        .str.replace(" ", "_")
    )
    return df

merged_df = standardize_columns(merged_df)
```

```python
top_batters_report = (
    merged_df.groupby('batter')['batsman_runs']
    .sum()
    .reset_index()
    .rename(columns={
        'batter': 'player_name',
        'batsman_runs': 'total_runs'
    })
    .sort_values(by='total_runs', ascending=False)
    .head(10)
)
```

```python
bowler_report = (
    merged_df.groupby('bowler')
    .agg(
        total_runs=('total_runs', 'sum'),
        balls_bowled=('ball', 'count')
    )
    .reset_index()
)

bowler_report['overs'] = bowler_report['balls_bowled'] / 6
bowler_report['economy'] = bowler_report['total_runs'] / bowler_report['overs']

bowler_report = (
    bowler_report[['bowler', 'economy']]
    .rename(columns={'bowler': 'player_name'})
    .sort_values(by='economy')
    .head(10)
)
```

```python
runs_per_match_report = (
    merged_df.groupby('match_id')['total_runs']
    .sum()
    .reset_index()
    .rename(columns={'total_runs': 'match_total_runs'})
    .sort_values(by='match_total_runs', ascending=False)
)
```

```python
venue_report = (
    merged_df.groupby('venue')
    .agg(
        total_matches=('match_id', 'nunique'),
        avg_runs=('total_runs', 'mean')
    )
    .reset_index()
    .sort_values(by='avg_runs', ascending=False)
)
```

```python
strike_rate_report = (
    merged_df.groupby('batter')
    .agg(
        total_runs=('batsman_runs', 'sum'),
        balls_faced=('ball', 'count')
    )
    .reset_index()
)

strike_rate_report['strike_rate'] = (
    strike_rate_report['total_runs'] /
    strike_rate_report['balls_faced'] * 100
)

strike_rate_report = strike_rate_report.sort_values(
    by='strike_rate', ascending=False
)
```

```python
bowler_report['economy'] = bowler_report['economy'].round(2)
strike_rate_report['strike_rate'] = strike_rate_report['strike_rate'].round(2)
venue_report['avg_runs'] = venue_report['avg_runs'].round(2)
```

```python
top_batters_report = top_batters_report[['player_name', 'total_runs']]
```

```python
import os

output_dir = "output"
os.makedirs(output_dir, exist_ok=True)
```

```python
team_scores = (
    runs_per_team_match
    .rename(columns={'batting_team': 'team'})
    .sort_values(['match_id', 'total_runs'], ascending=[True, False])
)

death_overs_report = (
    death_runs_team
    .reset_index()
    .rename(columns={
        'batting_team': 'team',
        'total_runs': 'death_over_runs'
    })
)
```

```python
runs_per_match_report.to_csv(f"{output_dir}/runs_per_match.csv", index=False)
top_batters_report.to_csv(f"{output_dir}/top_batters.csv", index=False)
strike_rate_report.to_csv(f"{output_dir}/strike_rate.csv", index=False)
bowler_report.to_csv(f"{output_dir}/economy.csv", index=False)
team_scores.to_csv(f"{output_dir}/team_scores.csv", index=False)
death_overs_report.to_csv(f"{output_dir}/death_overs.csv", index=False)
```

```python
from openpyxl.utils import get_column_letter

with pd.ExcelWriter(f"{output_dir}/ipl_analysis.xlsx", engine="openpyxl") as writer:
    def write_sheet(df, name):
        df.to_excel(writer, sheet_name=name, index=False)
        worksheet = writer.sheets[name]

        for i, col in enumerate(df.columns, start=1):
            max_len = max(df[col].astype(str).map(len).max(), len(col))
            worksheet.column_dimensions[get_column_letter(i)].width = max_len + 2

    write_sheet(runs_per_match_report, "Runs per Match")
    write_sheet(top_batters_report, "Top Batters")
    write_sheet(strike_rate_report, "Strike Rate")
    write_sheet(bowler_report, "Economy")
    write_sheet(team_scores, "Team Scores")
    write_sheet(death_overs_report, "Death Overs")
```
